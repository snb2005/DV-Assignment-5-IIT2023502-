```python
!pip install yfinance pandas

import pandas as pd
import yfinance as yf
from google.colab import files
import io

# STEP 1: UPLOAD YOUR MANUAL CSV
print("Please upload your manually created CSV file (iran_conflict_engineered_features.csv):")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Load it into Pandas
df_geo = pd.read_csv(io.BytesIO(uploaded[filename]))
df_geo['Date'] = pd.to_datetime(df_geo['Date'])
df_geo.set_index('Date', inplace=True)


# STEP 2: FETCH REAL YAHOO FINANCE DATA
print(f"\nFetching live Yahoo Finance data...")

# Tickers: BZ=F (Brent Crude), ^GSPC (S&P 500), GC=F (Gold), ^VIX (Volatility Index)
tickers = ["BZ=F", "^GSPC", "GC=F", "^VIX"]

# We pull up to April 12th to ensure we capture the April 10th close
fin_data = yf.download(tickers, start="2026-03-01", end="2026-04-12")['Close']

# yfinance alphabetizes columns by default: BZ=F, GC=F, ^GSPC, ^VIX
# Rename columns to match the assignment schema
fin_data.columns = ['Oil_Price', 'Gold', 'Stock_Index', 'Market_Volatility_Measure']


# STEP 3: MERGE & HANDLE WEEKENDS
print("Merging your CSV with the financial data...")

# Join based on the Date index
final_df = df_geo.join(fin_data, how='left')

# Handle Weekends: Forward fill Friday's closing prices into Saturday/Sunday
cols_to_fill = ['Oil_Price', 'Gold', 'Stock_Index', 'Market_Volatility_Measure']
final_df[cols_to_fill] = final_df[cols_to_fill].ffill().bfill()

# Clean up the decimals
final_df['Oil_Price'] = final_df['Oil_Price'].round(2)
final_df['Gold'] = final_df['Gold'].round(2)
final_df['Stock_Index'] = final_df['Stock_Index'].round(2)
final_df['Market_Volatility_Measure'] = final_df['Market_Volatility_Measure'].round(2)


# STEP 4: EXPORT FINAL DATASET
output_filename = "final_master_dataset.csv"
final_df.to_csv(output_filename)

# Automatically trigger the download back to your Ubuntu machine
files.download(output_filename)

print(f"\nSuccess! All engineered features included. Merged dataset saved as '{output_filename}'")
display(final_df.head())
```
